### Building the model

```python
transactions = []
for i in range(len(df2)):
    row = df2.loc[i].dropna().to_list()
    transactions.append(row)
```

```python
transactions
```

```python
## Creating the rules
from apyori import apriori
rules = apriori(transactions=transactions,min_support=0.004,min_confidence=0.2,min_lift=3,min_length=0,max_length=2)
results = list(rules)
```

```python
results
```

```python
tuple(results[0][0])
```

```python
results[0][1]
```

```python
results[0][2]
```

```python
def extract(results):
    ## Setting the lists
    item_sets = []
    support = []
    antecedent = []
    consquent = []
    confidence = []
    lift = []
    ## loop
    for result in results:
        for subset in result[2]:
            item_sets.append(tuple(result[0]))
            support.append(result[1])
            antecedent.append(subset[0])
            consquent.append(subset[1])
            confidence.append(subset[2])
            lift.append(subset[3])
    return list(zip(item_sets,support,antecedent,consquent,confidence,lift))
```

```python
results_df = pd.DataFrame(extract(results),columns=['Item_set','Support','Antecedent','Consquent','Confidence','Lift'])
results_df.head(4)
```
